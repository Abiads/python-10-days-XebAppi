```python
import time

# Original logger decorator
def logger_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"LOG: Function '{func.__name__}' started")
        result = func(*args, **kwargs)
        print(f"LOG: Function '{func.__name__}' ended")
        return result
    return wrapper

# New timer decorator
def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"TIME: Function '{func.__name__}' took {end_time - start_time:.4f} seconds")
        return result
    return wrapper

# Apply both decorators
@timer_decorator
@logger_decorator
def multiply(a, b):
    # Simulate some work
    time.sleep(0.1)
    return a * b

@logger_decorator
def greet(name):
    return f"Hello, {name}!"

print("\n--- Decorators Examples ---")
print(multiply(5, 4))
print(greet("Alice"))
```

```python
# Original number generator
def number_generator(n):
    for i in range(n):
        yield i

# New Fibonacci sequence generator
def fibonacci_generator(limit):
    a, b = 0, 1
    count = 0
    while count < limit:
        yield a
        a, b = b, a + b
        count += 1

# New infinite sequence generator
def infinite_sequence():
    num = 0
    while True:
        yield num
        num += 1

print("\n--- Generators Examples ---")
print("Numbers up to 5:")
gen_nums = number_generator(5)
for num in gen_nums:
    print(num)

print("Fibonacci sequence up to 10 terms:")
gen_fib = fibonacci_generator(10)
for fib_num in gen_fib:
    print(fib_num)

print("First 3 terms of infinite sequence:")
gen_inf = infinite_sequence()
for _ in range(3):
    print(next(gen_inf))
```

```python
print("\n--- Exception Handling Examples ---")

# Example 1: Original try-except-finally
try:
    # x = int(input("Enter an integer (original example): "))
    x = 0 # For demonstration, avoiding input interaction in automated execution
    y = 10 / x
    print(f"Result of 10 / {x}: {y}")
except ValueError:
    print("Original Example: Invalid input - Please enter an integer.")
except ZeroDivisionError:
    print("Original Example: Cannot divide by zero.")
finally:
    print("Original Example: Execution done.")

# Example 2: File handling with FileNotFoundError and general Exception
try:
    with open("non_existent_file.txt", "r") as f:
        content = f.read()
        print(content)
except FileNotFoundError:
    print("Extended Example: Error - The file was not found.")
except Exception as e:
    print(f"Extended Example: An unexpected error occurred: {e}")
finally:
    print("Extended Example: File operation attempt completed.")

# Example 3: Nested try-except blocks
def divide_and_convert(num_str, divisor_str):
    try:
        num = float(num_str)
        try:
            divisor = float(divisor_str)
            result = num / divisor
            print(f"Nested Example: {num} / {divisor} = {result}")
        except ZeroDivisionError:
            print("Nested Example: Inner block - Cannot divide by zero.")
    except ValueError:
        print("Nested Example: Outer block - Invalid number format.")

divide_and_convert("10", "2")
divide_and_convert("abc", "2")
divide_and_convert("10", "0")
```

```python
print("\n--- Custom Exception Examples ---")

# Original custom exception
class NegativeNumberError(Exception):
    """Raised when an operation is performed on a negative number where it's not allowed."""
    pass

def check_number(n):
    if n < 0:
        raise NegativeNumberError("Original Example: Negative numbers are not allowed.")
    return n

# New custom exception for out-of-range values
class OutOfRangeError(Exception):
    """Raised when a value is outside a specified valid range."""
    def __init__(self, value, lower_bound, upper_bound, message="Value is out of valid range"):
        self.value = value
        self.lower_bound = lower_bound
        self.upper_bound = upper_bound
        self.message = f"{message}: {value} (expected between {lower_bound} and {upper_bound})"
        super().__init__(self.message)

def check_range(value, lower, upper):
    if not (lower <= value <= upper):
        raise OutOfRangeError(value, lower, upper, "Extended Example: Value must be within specified bounds")
    return value

# Demonstrate original custom exception
try:
    print(f"Checking 10: {check_number(10)}")
    print(f"Checking -5: {check_number(-5)}")
except NegativeNumberError as e:
    print(e)

# Demonstrate new custom exception
try:
    print(f"Checking value 5 in range [1, 10]: {check_range(5, 1, 10)}")
    print(f"Checking value 0 in range [1, 10]: {check_range(0, 1, 10)}")
except OutOfRangeError as e:
    print(e)

try:
    print(f"Checking value 15 in range [1, 10]: {check_range(15, 1, 10)}")
except OutOfRangeError as e:
    print(e)
```

```python
import unittest

print("\n--- Unit Testing Examples ---")

# Function to be tested (original)
def multiply(a, b):
    return a * b

# New function to be tested
def divide(a, b):
    if b == 0:
        raise ValueError("Cannot divide by zero")
    return a / b

# Original test class with extended cases
class TestMath(unittest.TestCase):

    def test_multiply_positive(self):
        # Test with positive numbers
        self.assertEqual(multiply(2, 3), 6)
        self.assertEqual(multiply(10, 0), 0)

    def test_multiply_negative(self):
        # Test with negative numbers and mixed signs
        self.assertEqual(multiply(-1, 3), -3)
        self.assertEqual(multiply(-2, -4), 8)

    def test_multiply_float(self):
        # Test with floating-point numbers
        self.assertAlmostEqual(multiply(2.5, 2), 5.0)
        self.assertAlmostEqual(multiply(0.1, 0.2), 0.02)

    def test_divide_positive(self):
        # Test division with positive numbers
        self.assertEqual(divide(10, 2), 5)
        self.assertEqual(divide(5, 2), 2.5)

    def test_divide_negative(self):
        # Test division with negative numbers
        self.assertEqual(divide(-10, 2), -5)
        self.assertEqual(divide(10, -2), -5)
        self.assertEqual(divide(-10, -2), 5)

    def test_divide_by_zero(self):
        # Test division by zero should raise ValueError
        with self.assertRaises(ValueError):
            divide(10, 0)

# This part ensures that tests run when the script is executed directly
if __name__ == '__main__':
    # Use unittest.main() to discover and run all tests in the current file
    # For running in Colab, this will execute the tests and print results.
    # Using verbosity=2 for more detailed output.
    unittest.main(argv=['first-arg-is-ignored'], exit=False, verbosity=2)
```
