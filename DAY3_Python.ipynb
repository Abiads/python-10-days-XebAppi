```python
# Daily Report
# Date: 23 March 2025

# Work Done
# • Explored basic error handling using try and except blocks to manage runtime errors and prevent crashes
# • Implemented exception handling in programs to handle invalid inputs
# • Tested error scenarios like division by zero and invalid input
# • Applied exception handling to improve program reliability
# • Debugged code using try-except blocks

# Learning / Concepts Covered
# • try block for risky code
# • except block for handling errors
# • Common exceptions: ZeroDivisionError, ValueError, TypeError

# Observation
# • Improved understanding of runtime error handling
# • Learned to write stable programs
# • Gained confidence in debugging
# • Understood importance of validation
# • Built base for advanced error handling

# BASIC TRY-EXCEPT EXAMPLE

try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print("Result:", result)

except ZeroDivisionError:
    print("Error: Cannot divide by zero")

except ValueError:
    print("Error: Invalid input, please enter a number")
```

```python
# MULTIPLE EXCEPTIONS + ELSE + FINALLY

try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    result = num1 / num2

except ZeroDivisionError:
    print("Error: Division by zero is not allowed")

except ValueError:
    print("Error: Please enter valid integers")

except Exception as e:
    print("Unexpected Error:", e)

else:
    print("Result:", result)

finally:
    print("Execution completed (this block always runs)")
```

```python
# LOOP WITH ERROR HANDLING (UNTIL VALID INPUT)

while True:
    try:
        value = int(input("Enter a positive number: "))

        if value > 0:
            print("Valid Input:", value)
            break
        else:
            print("Number must be positive")

    except ValueError:
        print("Invalid input! Please enter a number")
```

```python
# CUSTOM EXCEPTION

class NegativeNumberError(Exception):
    pass

try:
    num = int(input("Enter a positive number: "))

    if num < 0:
        raise NegativeNumberError("Negative numbers are not allowed")

    print("Valid Number:", num)

except NegativeNumberError as e:
    print("Custom Error:", e)

except ValueError:
    print("Error: Please enter a valid integer")
```

```python
# FILE HANDLING WITH ERROR HANDLING

try:
    file = open("sample.txt", "r")
    content = file.read()
    print("File Content:\n", content)

except FileNotFoundError:
    print("Error: File not found. Please check the file name")

except IOError:
    print("Error: Issue while reading the file")

finally:
    try:
        file.close()
        print("File closed successfully")
    except:
        print("File was not opened")
```

```python
# USING CONTEXT MANAGER (BEST PRACTICE)

try:
    with open("sample.txt", "r") as file:
        content = file.read()
        print("File Content:\n", content)

except FileNotFoundError:
    print("Error: File not found")

except Exception as e:
    print("Unexpected Error:", e)
```
