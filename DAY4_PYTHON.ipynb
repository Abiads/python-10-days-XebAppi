```python
from abc import ABC, abstractmethod
import math

# 2. Define an abstract base class named Shape that inherits from ABC.
class Shape(ABC):
    # 3. Within the Shape class, define two abstract methods: area() and perimeter().
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

# 4. Define a concrete class named Circle that inherits from Shape.
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    # 5. Implement the area() and perimeter() methods within the Circle class.
    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

# 6. Create an instance of the Circle class, for example, with a radius of 5.
circle_instance = Circle(radius=5)

# 7. Call the area() and perimeter() methods on the Circle instance and print their results.
print(f"Circle Area: {circle_instance.area():.2f}")
print(f"Circle Perimeter: {circle_instance.perimeter():.2f}")
```

```python
import math

# 1. Define a class named Engine with an __init__ method that takes horsepower and fuel_type as arguments and stores them as attributes. Add a method named start() that prints a message indicating the engine is starting.
class Engine:
    def __init__(self, horsepower, fuel_type):
        self.horsepower = horsepower
        self.fuel_type = fuel_type

    def start(self):
        print(f"The {self.fuel_type} engine with {self.horsepower} HP is starting.")

# 2. Define a class named Car with an __init__ method that takes make, model, and engine_type (e.g., 'V6', 'Electric') as arguments. Inside the Car's __init__, create an instance of the Engine class based on the engine_type (e.g., if 'V6', instantiate Engine(horsepower=300, fuel_type='Gasoline')). Store this Engine instance as an attribute of the Car (e.g., self.engine).
class Car:
    def __init__(self, make, model, engine_type):
        self.make = make
        self.model = model
        if engine_type == 'V6':
            self.engine = Engine(horsepower=300, fuel_type='Gasoline')
        elif engine_type == 'Electric':
            self.engine = Engine(horsepower=250, fuel_type='Electric')
        else:
            self.engine = Engine(horsepower=150, fuel_type='Unknown') # Default or custom engine

    # 3. Add a method to the Car class named drive() that prints a message indicating the car is driving and then calls the start() method of its engine attribute.
    def drive(self):
        print(f"The {self.make} {self.model} is driving.")
        self.engine.start()

# 4. Create an instance of the Car class, for example, a 'Toyota Camry' with a 'V6' engine.
my_car = Car(make='Toyota', model='Camry', engine_type='V6')

# 5. Call the drive() method on the Car instance to demonstrate the composition.
my_car.drive()

electric_car = Car(make='Tesla', model='Model 3', engine_type='Electric')
electric_car.drive()
```

```python
import math

# 1. Define a base class named Animal with an __init__ method that takes a name and stores it as an attribute. Add a method named speak() that prints a generic message like 'The animal makes a sound.'.
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"The {self.name} makes a sound.")

# 2. Define a derived class named Dog that inherits from Animal. The Dog class's __init__ method should call the parent class's __init__ using super().__init__(name) and also add a breed attribute.
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    # 3. Override the speak() method in the Dog class to print a specific message for a dog, such as 'Woof! Woof!'.
    def speak(self):
        print(f"{self.name}, the {self.breed} says: Woof! Woof!")

# 4. Define another derived class named Cat that also inherits from Animal. The Cat class's __init__ method should similarly call the parent's __init__ and add a color attribute.
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name)
        self.color = color

    # 5. Override the speak() method in the Cat class to print a specific message for a cat, such as 'Meow.'.
    def speak(self):
        print(f"{self.name}, the {self.color} cat, says: Meow.")

# 6. Create an instance of the Animal class, a Dog class, and a Cat class.
animal_instance = Animal(name="Generic Animal")
dog_instance = Dog(name="Buddy", breed="Golden Retriever")
cat_instance = Cat(name="Whiskers", color="Orange")

# 7. Call the speak() method on each instance to demonstrate method overriding.
print("Demonstrating Method Overriding:")
animal_instance.speak()
dog_instance.speak()
cat_instance.speak()
```

### Q&A
The key OOP concepts covered include Abstraction, Composition, Method Overriding, and Multiple Inheritance.
*   **Abstraction** is useful for defining a common interface or blueprint for a set of classes, ensuring that all subclasses implement certain methods. It helps in managing complexity by hiding implementation details and showing only essential features.
*   **Composition** is useful for building complex objects from simpler ones, representing a "has-a" relationship. It promotes code reuse and flexibility by allowing independent development and testing of component parts.
*   **Method Overriding** is useful when a subclass needs to provide a specific implementation for a method that is already defined in its parent class. This allows for specialized behavior while maintaining a common interface defined by the parent.
*   **Multiple Inheritance** (though generally used cautiously due to complexity) is useful when a class needs to inherit functionalities from multiple distinct parent classes, combining their behaviors into a single entity.

```python
class Father:
    def __init__(self, name):
        self.name = name
        self.hobby_name = "fishing"
        print(f"Father {self.name} is initialized.")

    def hobby(self):
        print(f"{self.name} enjoys {self.hobby_name}.")

class Mother:
    def __init__(self, name):
        self.name = name
        self.skill_name = "cooking"
        print(f"Mother {self.name} is initialized.")

    def skill(self):
        print(f"{self.name} is skilled at {self.skill_name}.")

class Child(Father, Mother):
    def __init__(self, name, father_name, mother_name):
        self.name = name # Child's own name
        # Initialize parent classes explicitly for clarity in multiple inheritance
        Father.__init__(self, father_name)
        Mother.__init__(self, mother_name)
        print(f"Child {self.name} is initialized, with father {father_name} and mother {mother_name}.")

    def play(self):
        print(f"Child {self.name} is playing.")

# Create an instance of the Child class
child_instance = Child(name="Alice", father_name="Bob", mother_name="Carol")

# Call the hobby() method (from Father) and the skill() method (from Mother) on the Child instance
child_instance.hobby()
child_instance.skill()

# Call the play() method on the Child instance
child_instance.play()

# Print the Method Resolution Order (MRO) for the Child class
print("\nMethod Resolution Order (MRO) for Child:")
print(Child.mro())
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"Person {self.name} is initialized with age {self.age}.")

    def get_person_details(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Employee(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)  # Call the parent class's constructor
        self.employee_id = employee_id
        print(f"Employee {self.name} (ID: {self.employee_id}) is initialized.")

    def get_employee_details(self):
        print(f"Name: {self.name}, Age: {self.age}, Employee ID: {self.employee_id}")

# Create an instance of the Employee class
employee_instance = Employee(name='Alice', age=30, employee_id='EMP001')

# Call the get_employee_details() method on the Employee instance
print("\nDemonstrating Employee details:")
employee_instance.get_employee_details()
```

```python
class Shape:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        print('Calculating area of a shape.')
        return 0

class Rectangle(Shape):
    def __init__(self, width, height):
        super().__init__(width, height)

    def area(self):
        return self.width * self.height

# Create an instance of the Shape class
shape_instance = Shape(10, 5)

# Create an instance of the Rectangle class
rectangle_instance = Rectangle(10, 5)

# Call the area() method on both instances and print their results
print(f"Shape Area: {shape_instance.area()}")
print(f"Rectangle Area: {rectangle_instance.area()}")
```

```python
class College:
    # 2. Define a class variable named total_colleges
    total_colleges = 0

    # 3. Implement the __init__ method for the College class
    def __init__(self, name, city):
        self.name = name
        self.city = city
        # Increment the total_colleges class variable
        College.total_colleges += 1
        print(f"College instance created: {self.name} in {self.city}. Total colleges: {College.total_colleges}")

    # 4. Define a class method named create_from_string
    @classmethod
    def create_from_string(cls, college_info_string):
        # 6. Parse the college_info_string
        name, city = college_info_string.split('-')
        # Use cls(name, city) to create and return a new College instance
        return cls(name, city)

# 7. Create two instances of the College class using the standard constructor
print("\n--- Creating instances with standard constructor ---")
college1 = College('SRIT', 'MNIT')
college2 = College('IIT', 'UIT')

# 8. Create a third instance using the create_from_string class method
print("\n--- Creating instance with class method ---")
college3 = College.create_from_string('Jabalpur-MP')

# 9. Print the name and city for each instance
print("\n--- College Details ---")
print(f"College 1: {college1.name}, City: {college1.city}")
print(f"College 2: {college2.name}, City: {college2.city}")
print(f"College 3: {college3.name}, City: {college3.city}")

# 10. Print the total_colleges class variable
print("\n--- Total Colleges ---")
print(f"Total colleges created (via class): {College.total_colleges}")
print(f"Total colleges created (via instance college1): {college1.total_colleges}")
print(f"Total colleges created (via instance college3): {college3.total_colleges}")
```

```python
import shutil
import os

# 2. Create a dummy file named source_file.txt with some content in the current working directory.
source_file_name = 'source_file.txt'
with open(source_file_name, 'w') as f:
    f.write('This is a dummy file for shutil demonstration.')
print(f"Created: {source_file_name}")

# 3. Create a directory named destination_copy_dir.
dest_copy_dir = 'destination_copy_dir'
os.makedirs(dest_copy_dir, exist_ok=True)
print(f"Created directory: {dest_copy_dir}")

# 4. Use shutil.copy() to copy source_file.txt into destination_copy_dir. Print a confirmation message.
shutil.copy(source_file_name, dest_copy_dir)
print(f"Copied '{source_file_name}' to '{dest_copy_dir}'")

# 5. Create a directory named destination_move_dir.
dest_move_dir = 'destination_move_dir'
os.makedirs(dest_move_dir, exist_ok=True)
print(f"Created directory: {dest_move_dir}")

# 6. Use shutil.move() to move source_file.txt (from its original location, or its current location if moved in previous steps) into destination_move_dir.
# Note: After the copy operation, source_file.txt still exists in the current directory.
shutil.move(source_file_name, dest_move_dir)
print(f"Moved '{source_file_name}' to '{dest_move_dir}'")

# 7. Create another dummy directory named dir_to_remove and create a file inside it (e.g., dir_to_remove/test.txt) to demonstrate rmtree on a non-empty directory.
dir_to_remove = 'dir_to_remove'
os.makedirs(dir_to_remove, exist_ok=True)
with open(os.path.join(dir_to_remove, 'test.txt'), 'w') as f:
    f.write('This file is inside a directory to be removed.')
print(f"Created directory '{dir_to_remove}' with 'test.txt' inside.")

# 8. Use shutil.rmtree() to remove dir_to_remove. Print a confirmation message.
shutil.rmtree(dir_to_remove)
print(f"Removed directory '{dir_to_remove}' using shutil.rmtree()")

# 9. Clean up any remaining created files or directories using os.remove() or os.rmdir() if necessary (e.g., destination_copy_dir and destination_move_dir).
# Ensure to check if they exist before attempting to remove them.
# The original source_file.txt has been moved.

# Clean up destination_copy_dir
if os.path.exists(dest_copy_dir):
    # The copied file is inside, so rmtree is appropriate for non-empty directory
    shutil.rmtree(dest_copy_dir)
    print(f"Cleaned up directory: {dest_copy_dir}")

# Clean up destination_move_dir
if os.path.exists(dest_move_dir):
    # The moved file is inside, so rmtree is appropriate for non-empty directory
    shutil.rmtree(dest_move_dir)
    print(f"Cleaned up directory: {dest_move_dir}")

print("Shutil demonstration complete and cleanup performed.")
```

```python
import json
import os

# 2. Create a Python dictionary with some sample data.
data = {
    'name': 'Alice',
    'age': 30,
    'isStudent': False,
    'courses': [
        {'title': 'Math', 'credits': 3},
        {'title': 'Science', 'credits': 4}
    ]
}

print("Original Python Dictionary:")
print(data)

# 3. Use json.dumps() to convert the Python dictionary to a JSON formatted string and print the result.
json_string = json.dumps(data, indent=4)
print("\nPython Dictionary converted to JSON string:")
print(json_string)

# 4. Use json.loads() to convert the JSON string back to a Python dictionary and print the result.
decoded_data = json.loads(json_string)
print("\nJSON string decoded back to Python Dictionary:")
print(decoded_data)

# 5. Define a filename for a dummy JSON file.
file_name = 'data.json'

# 6. Use open() in write mode ('w') and json.dump() to write the Python dictionary to the dummy JSON file.
with open(file_name, 'w') as f:
    json.dump(data, f, indent=4)
print(f"\nPython Dictionary saved to '{file_name}'.")

# 7. Use open() in read mode ('r') and json.load() to read the JSON data from the dummy JSON file back into a Python variable and print the result.
with open(file_name, 'r') as f:
    loaded_data = json.load(f)
print(f"\nData loaded from '{file_name}':")
print(loaded_data)

# 8. Remove the dummy JSON file using os.remove() to clean up the environment.
os.remove(file_name)
print(f"\nCleaned up: Removed '{file_name}'.")
```
