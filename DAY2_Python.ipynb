```python
# TUPLE OPERATIONS
coordinates = (10, 20, 30)

# Access elements
print("First Element:", coordinates[0])

# Tuple unpacking
x, y, z = coordinates
print("Unpacked:", x, y, z)

# Attempt modification (will cause error if uncommented)
# coordinates[0] = 100
```

```python
# DICTIONARY OPERATIONS
student = {
    "name": "Abhay",
    "age": 21,
    "course": "CSE"
}

# Access value
print("Name:", student["name"])

# Add new key
student["grade"] = "A"

# Update value
student["age"] = 22

# Iterate
for key, value in student.items():
    print(key, ":", value)
```

```python
# SET OPERATIONS
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Union
print("Union:", set1.union(set2))

# Intersection
print("Intersection:", set1.intersection(set2))

# Difference
print("Difference:", set1.difference(set2))

# Add element
set1.add(7)
print("Updated Set1:", set1)
```
