## Part 3: Operators




```python
num = 10
print(f"Initial value of num: {num}")

num += 5  # num = num + 5
print(f"After num += 5: {num}")

num -= 3  # num = num - 3
print(f"After num -= 3: {num}")

num *= 2  # num = num * 2
print(f"After num *= 2: {num}")

num /= 4  # num = num / 4
print(f"After num /= 4: {num}")

num %= 3  # num = num % 3
print(f"After num %= 3: {num}")

num = 2 # Reset num for exponentiation example
num **= 3 # num = num ** 3
print(f"After num **= 3: {num}")

num = 15 # Reset num for floor division example
num //= 4 # num = num // 4
print(f"After num //= 4: {num}")
```

```python
a = 10
b = 3

print(f"a = {a}, b = {b}")

# Addition
print(f"Addition (a + b): {a + b}")

# Subtraction
print(f"Subtraction (a - b): {a - b}")

# Multiplication
print(f"Multiplication (a * b): {a * b}")

# Division
print(f"Division (a / b): {a / b}")

# Floor Division
print(f"Floor Division (a // b): {a // b}")

# Modulo
print(f"Modulo (a % b): {a % b}")

# Exponentiation
print(f"Exponentiation (a ** b): {a ** b}")
```

```python
x = 15
y = 7
z = 15

print(f"x = {x}, y = {y}, z = {z}")

# Equal to
print(f"Equal to (x == y): {x == y}")
print(f"Equal to (x == z): {x == z}")

# Not equal to
print(f"Not equal to (x != y): {x != y}")
print(f"Not equal to (x != z): {x != z}")

# Greater than
print(f"Greater than (x > y): {x > y}")
print(f"Greater than (y > x): {y > x}")

# Less than
print(f"Less than (x < y): {x < y}")
print(f"Less than (y < x): {y < x}")

# Greater than or equal to
print(f"Greater than or equal to (x >= y): {x >= y}")
print(f"Greater than or equal to (x >= z): {x >= z}")

# Less than or equal to
print(f"Less than or equal to (x <= y): {x <= y}")
print(f"Less than or equal to (x <= z): {x <= z}")
```

```python
is_sunny = True
is_warm = False
is_weekend = True

print(f"is_sunny = {is_sunny}, is_warm = {is_warm}, is_weekend = {is_weekend}")

# Logical AND
print(f"Logical AND (is_sunny and is_warm): {is_sunny and is_warm}")
print(f"Logical AND (is_sunny and is_weekend): {is_sunny and is_weekend}")

# Logical OR
print(f"Logical OR (is_sunny or is_warm): {is_sunny or is_warm}")
print(f"Logical OR (is_warm or False): {is_warm or False}")

# Logical NOT
print(f"Logical NOT (not is_sunny): {not is_sunny}")
print(f"Logical NOT (not is_warm): {not is_warm}")
```

### Practice Program 5: Temperature Converter
This program will convert a temperature from Celsius to Fahrenheit, or vice versa, based on user input.

**Reasoning**:
Now that the description for 'Practice Program 5: Temperature Converter' is in place, I need to add its code implementation to the notebook.



```python
temperature_str = input("Enter the temperature value: ")
temperature = float(temperature_str)

unit = input("Enter the unit of temperature (C for Celsius, F for Fahrenheit): ").upper()

if unit == 'C':
    fahrenheit = (temperature * 9/5) + 32
    print(f"{temperature}°C is equal to {fahrenheit}°F.")
elif unit == 'F':
    celsius = (temperature - 32) * 5/9
    print(f"{temperature}°F is equal to {celsius}°C.")
else:
    print("Invalid unit entered. Please enter 'C' or 'F'.")
```

## Jupyter Notebook Shortcuts

Jupyter Notebooks offer several shortcuts to enhance productivity. Here are some common ones:

### Command Mode (Cell not being edited)
*   `A`: Insert cell above
*   `B`: Insert cell below
*   `X`: Cut selected cell(s)
*   `C`: Copy selected cell(s)
*   `V`: Paste cell(s) below
*   `DD`: Delete selected cell(s)
*   `Z`: Undo cell deletion
*   `Y`: Change cell to Code
*   `M`: Change cell to Markdown
*   `R`: Change cell to Raw NBConvert
*   `L`: Toggle line numbers
*   `Shift + Up/Down`: Select multiple cells
*   `Shift + M`: Merge selected cells
*   `Enter`: Enter edit mode
*   `P`: Open the command palette

### Edit Mode (Cell being edited)
*   `Esc`: Enter command mode
*   `Ctrl + Enter`: Run selected cell
*   `Shift + Enter`: Run selected cell and insert new cell below (or select cell below)
*   `Alt + Enter`: Run selected cell and insert new cell below
*   `Ctrl + S`: Save and Checkpoint
*   `Ctrl + Shift + F`: Open Find and Replace
*   `Ctrl + Shift + P`: Open command palette
*   `Tab`: Code completion or indent
*   `Shift + Tab`: Tooltip (documentation for selected object)
*   `Ctrl + ]`: Indent
*   `Ctrl + [`: Dedent
*   `Ctrl + A`: Select all
*   `Ctrl + Z`: Undo
*   `Ctrl + Y`: Redo
